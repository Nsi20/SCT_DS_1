```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(age_gender_data['Year'], age_gender_data['Value'], color='skyblue')

plt.title('Population Ages 70-74, Male', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(age_gender_data['Value'], bins=10, color='skyblue', edgecolor='black')

plt.title('Distribution of Population Ages 70-74, Male', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.tight_layout()

plt.show()
```

```python
female_data = df[df['Indicator Name'] == 'Population ages 70-74, female'].copy()
female_data['Value'] = pd.to_numeric(female_data['Value'], errors='coerce')

```

# **Comparing Across Groups**

```python
male_data = df[df['Indicator Name'] == 'Population ages 70-74, male'].copy()
female_data = df[df['Indicator Name'] == 'Population ages 70-74, female'].copy()

male_data['Value'] = pd.to_numeric(male_data['Value'], errors='coerce')
female_data['Value'] = pd.to_numeric(female_data['Value'], errors='coerce')

comparison_data = pd.merge(
    male_data[['Year', 'Value']],
    female_data[['Year', 'Value']],
    on='Year',
    suffixes=('_male', '_female')
)

# Plot comparison
plt.figure(figsize=(10, 6))
plt.plot(comparison_data['Year'], comparison_data['Value_male'], label='Male', color='blue')
plt.plot(comparison_data['Year'], comparison_data['Value_female'], label='Female', color='pink')
plt.title('Population Ages 70-74: Male vs. Female', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Analyzing Trends Over Time**

```python
yearly_population = age_gender_data.groupby('Year')['Value'].sum()


plt.figure(figsize=(10, 6))
plt.plot(yearly_population.index, yearly_population.values, marker='o', color='green')
plt.title('Total Population Trend for Ages 70-74, Male', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Combining Visualizations**

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(15, 12))

# Subplot 1: Bar chart for population trend
plt.subplot(2, 2, 1)
plt.bar(age_gender_data['Year'], age_gender_data['Value'], color='skyblue')
plt.title('Population Ages 70-74, Male')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)

# Subplot 2: Histogram for population distribution
plt.subplot(2, 2, 2)
plt.hist(age_gender_data['Value'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Population Ages 70-74, Male')
plt.xlabel('Population')
plt.ylabel('Frequency')

# Subplot 3: Line plot comparing male and female populations
plt.subplot(2, 2, 3)
plt.plot(comparison_data['Year'], comparison_data['Value_male'], label='Male', color='blue')
plt.plot(comparison_data['Year'], comparison_data['Value_female'], label='Female', color='pink')
plt.title('Population Ages 70-74: Male vs. Female')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()

# Subplot 4: Line plot for total population trend
yearly_population = age_gender_data.groupby('Year')['Value'].sum()
plt.subplot(2, 2, 4)
plt.plot(yearly_population.index, yearly_population.values, marker='o', color='green')
plt.title('Total Population Trend for Ages 70-74, Male')
plt.xlabel('Year')
plt.ylabel('Population')

plt.tight_layout()

plt.show()
```
